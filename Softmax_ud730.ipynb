```python
"""Softmax."""
## Nabot, date: 29-12-2017

scores = [2.0, 1.0, 0.2]
#scores = [.0, 1.0, 0.2]
print(scores)

import numpy as np
import math


def softmax(x):

    """Compute softmax values for each sets of scores in x."""
    pass  # TODO: Compute and return softmax(x)
    sum_denominator = 0
    powers = []
    for item in x:
        powers.append(math.e**item)
        sum_denominator = sum_denominator + powers[-1]
    for idx in range(len(x)):
        x[idx] = powers[idx]/sum_denominator
    return x

after_comput = softmax(scores)

print(after_comput)
print(after_comput[0]+after_comput[1]+after_comput[2])

# Plot softmax curves
import matplotlib.pyplot as plt
x = np.arange(-2.0, 6.0, 0.1)
scores = np.vstack([x, np.ones_like(x), 0.2 * np.ones_like(x)])
print(scores)
print(softmax(scores).T)

plt.plot(x, softmax(scores).T, linewidth=2)   
#plt.plot(x, softmax(1.0).T, linewidth=2)     
plt.show() 
```
